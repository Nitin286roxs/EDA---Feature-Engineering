# Imporing matplotlib, pandas and numpy


```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
%matplotlib
```

## Load train and test csv

```
train_df = pd.read_csv('/content/black_friday_train.csv')
test_df = pd.read_csv('/content/black_friday_test.csv')
```

```
train_df.head(2)
```

```
test_df.head(2)
```

## Combine train and test dataset

```
combined_df = pd.concat([train_df, test_df])
```

```
combined_df.head(2)
```

## Remove irrevent keys from dataset

```
combined_df.head(5)
```

```
combined_df.drop(['User_ID'],axis=1,inplace=True)
```

```
combined_df.head(5)
```

## Convert catagarical columns to numerical

### Handing Gender

```
combined_df['Gender']=combined_df['Gender'].map({'F':0,'M':1})
```

### Handling Age

```
combined_df['Age'].unique()
```

```
combined_df['Age']=combined_df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7 })
```

```
combined_df.head(5)
```

### Handling City Catagary

```
combined_df['City_Category'].unique()
```

```
df_cat=pd.get_dummies(combined_df['City_Category'], drop_first=True,dtype=int)
```

```
df_cat.head(5)
```

```
combined_df=pd.concat([combined_df,df_cat],axis=1)
```

```
combined_df.drop(['City_Category'],axis=1,inplace=True)
```

```
combined_df.head(5)
```

## Missing Value

```
combined_df.isnull().sum()
```

### focus on replacing missing values

```
## Replace the Product_Category_2	with mode
combined_df['Product_Category_2'].mode()
combined_df['Product_Category_2']=combined_df['Product_Category_2'].fillna(combined_df['Product_Category_2'].mode()[0])
```

```
train_df['Product_Category_2'].isnull().sum()
```

```
combined_df['Product_Category_3']=combined_df['Product_Category_3'].fillna(combined_df['Product_Category_3'].mode()[0])
```

```
combined_df['Product_Category_3'].isnull().sum()
```

```
combined_df['Stay_In_Current_City_Years'].unique()
```

```
combined_df['Stay_In_Current_City_Years']=combined_df['Stay_In_Current_City_Years'].str.replace('+','')
```

#

```
combined_df.info()
```

```
combined_df['Stay_In_Current_City_Years'] = combined_df['Stay_In_Current_City_Years'].astype(int)
```

```
combined_df.info()
```

# Feature Scaling

```
test_df = combined_df[combined_df['Purchase'].isnull()]
train_df = combined_df[~combined_df['Purchase'].isnull()]
```

```
train_df.head(5)
```

```
test_df.head(5)
```

# Create X_train, Y_train, X_test and Y_test

```
y_train = train_df['Purchase']
y_train.shape
x_train = train_df.drop(['Purchase'],axis=1)
x_train.shape
```

## split train set

```
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(x_train,y_train,test_size=0.33,random_state=42)
```

## Drop Product ID

```
from re import X
X_train.drop(['Product_ID'], axis=1, inplace=True)
X_test.drop(['Product_ID'], axis=1, inplace=True)
```

```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

#Train the model

```
# Build the Regression / Regressor models

from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn import linear_model
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
import xgboost as xgb
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn import metrics
```

```
a={'Model Name':[], 'R2_score':[]}
Results=pd.DataFrame(a)
Results.head()
```

```
# Create objects of Regression / Regressor models with default hyper-parameters
modelmlg = LinearRegression()
modeldcr = DecisionTreeRegressor()
modelbag = BaggingRegressor()
modelrfr = RandomForestRegressor()
modelSVR = SVR()
modelXGR = xgb.XGBRegressor()
modelKNN = KNeighborsRegressor(n_neighbors=5)
modelETR = ExtraTreesRegressor()
modelRE=Ridge()
modelLO=linear_model.Lasso(alpha=0.1)

# Evalution matrix for all the algorithms
MM = [modelmlg, modeldcr, modelrfr, modelKNN, modelETR, modelXGR, modelbag,modelRE,modelLO]
for models in MM:
    # Fit the model with train data
    models.fit(X_train, Y_train)
    # Predict the model with test data
    y_pred = models.predict(X_test)
    # Print the model name
    print('Model Name: ', models)
    #TODO Calculate R2-Score
    r2_score=round(metrics.r2_score(Y_test, y_pred),6)
    # Evaluation metrics for Regression analysis
    print('R2_score:', r2_score)

    new_row = [{"Model": models,\
               "R2_score": r2_score}]
    new_row_df = pd.DataFrame(new_row, columns=["Models", "R2_score"])
    Results = pd.concat([Results, new_row_df])
Results

```

# XGBRegressor models has higer r2-score, we will use XGBRegressor.
