# Import pands to load dataset

```python
import pandas as pd
```

# Upload train and test dataset file

```python
train_df = pd.read_excel("Data_Train.xlsx")
test_df = pd.read_excel("Test_set.xlsx")
```

# Concat train and test for preprocessing

```python
final_df = pd.concat([train_df, test_df])
```

```python
final_df.isnull().sum()
```

# We can drop row=9039

```python
final_df.drop([9039], axis=0, inplace=True)
final_df.isnull().sum()
```

# Split Date_of_Journey in differenent columns

```python
final_df["Journey_Day"] = final_df["Date_of_Journey"].str.split("/").str[0]
final_df["Journey_Month"] = final_df["Date_of_Journey"].str.split("/").str[1]
final_df["Journey_Year"] = final_df["Date_of_Journey"].str.split("/").str[2]
```

```python
final_df.drop(['Date_of_Journey'], axis=1, inplace=True)
```

```python
#final_df['Airline'].unique()
#final_df['Airline']=final_df['Airline'].map({'IndiGo': 1, 'Air India':2 , 'Jet Airways':3 , 'SpiceJet':4,
#       'Multiple carriers':5 , 'GoAir': 6, 'Vistara': 7, 'Air Asia': 8,
#       'Vistara Premium economy':9, 'Jet Airways Business': 10,
#       'Multiple carriers Premium economy':11, 'Trujet':12})
```

# drop routes and Additional_Info



```python
final_df.drop(['Route'], axis=1, inplace=True)
final_df.head(5)
```

#convert day, month and year into int

```python
final_df['Journey_Day']= final_df['Journey_Day'].astype(int)
final_df['Journey_Month']= final_df['Journey_Month'].astype(int)
final_df['Journey_Year']= final_df['Journey_Year'].astype(int)
```

# split  arrival_time coloumn

```python
final_df['Arrival_Time']=final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
final_df.head( 4)
```

```python
final_df["Arrival_hr"] = final_df['Arrival_Time'].str.split(":").str[0]
final_df["Arrival_min"] = final_df['Arrival_Time'].str.split(":").str[1]
final_df.drop(['Arrival_Time'], axis=1, inplace=True)
final_df.head(5)
```

```python
final_df['Arrival_hr']= final_df['Arrival_hr'].astype(int)
final_df['Arrival_min']= final_df['Arrival_min'].astype(int)
```

# split departure time into different columns

```python
final_df["Depart_hr"] = final_df['Dep_Time'].str.split(":").str[0]
final_df["Depart_min"] = final_df['Dep_Time'].str.split(":").str[1]
final_df.drop(['Dep_Time'], axis=1, inplace=True)
final_df["Depart_hr"]= final_df["Depart_hr"].astype(int)
final_df["Depart_min"]= final_df["Depart_min"].astype(int)
```

```python
final_df.info()
```

```python
final_df.head(5)
```

# Map Number of with 0,1,2,3

```python
final_df['Total_Stops'].unique()
```

```python
final_df['Total_Stops'] = final_df['Total_Stops'].map({'non-stop': 0, "2 stops": 2, "1 stop": 1, "3 stops": 3, "4 stops": 4})
```

```python
final_df.head(5)
#final_df.info()
```

# Convert duration into minutes

```python
final_df['Duration_hr']=final_df['Duration'].str.split(" ").str[0].str.split("h").str[0]
```

```python
final_df.head(5)
```

```python
final_df[final_df['Duration_hr']=='5m']
final_df.drop([6474], axis=0, inplace=True)
final_df.drop([2660], axis=0, inplace=True)
```

```python
final_df['Duration_hr']= final_df['Duration_hr'].astype(int)
```

```python
final_df.head(5)
```

```python
final_df['Duration_min']=final_df['Duration'].str.split(" ").str[1].str.split("m").str[0]
final_df['Duration_min']=final_df['Duration_min'].fillna(0)
final_df['Duration_min']=final_df['Duration_min'].astype(int)
```

```python
final_df.info()
```

```python
final_df['Duration']=final_df["Duration_hr"]*60 + final_df["Duration_min"]
```

```python
final_df.head(5)
```

```python
final_df.drop(['Duration_hr', 'Duration_min'], axis=1, inplace=True)
```

```python
final_df.head(5)
```

```python
final_df.head(5)
```

```python
X_train=final_df.drop(['Price'], axis=1)
y_train=final_df['Price']
```
